# Explore the medallion lake
Query DuckDB and the Parquet layers directly. Run the `medallion_hello_world` job in Dagster (http://<host>:3000) first to populate the layers.

```python
import os

import duckdb

con = duckdb.connect(os.environ.get("DUCKDB_PATH", "/app/data/warehouse.duckdb"), read_only=True)
con.sql("show all tables")
```

```python
# Bronze straight from Parquet
data_dir = os.environ.get("DATA_DIR", "/app/data")
con.sql(f"select * from read_parquet('{data_dir}/bronze/users.parquet') limit 10").df()
```

```python
# Silver (dbt view)
con.sql("select * from main_silver.stg_users limit 10").df()
```

```python
# Gold aggregate
con.sql("select * from main_gold.dim_users_by_city order by user_count desc").df()
```
