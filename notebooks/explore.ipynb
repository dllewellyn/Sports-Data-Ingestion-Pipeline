# Explore the medallion lake
Query the exported Parquet layers directly with DuckDB. Run a source job (e.g. `espn_ingestion`) in Dagster (http://<host>:3000) first to populate bronze + the gold exports.

> Notebooks read the **gold Parquet exports**, not the DuckLake catalog directly (dbt owns the catalog).

```python
import os

import duckdb

# In-memory connection; we read the Parquet files, not the DuckLake catalog.
data_dir = os.environ.get("DATA_DIR", "/app/data")
con = duckdb.connect()
con.sql(f"select * from glob('{data_dir}/gold/*.parquet')")
```

```python
# Bronze straight from Parquet (ESPN scoreboards)
con.sql(f"select * from read_parquet('{data_dir}/bronze/espn/**/*.parquet') limit 10").df()
```

```python
# Canonical matches (gold export of int_match)
path = f"{data_dir}/gold/canonical_match_export.parquet"
con.sql(f"select * from read_parquet('{path}') limit 10").df()
```

```python
# Completed matches mart
path = f"{data_dir}/gold/completed_matches_export.parquet"
con.sql(f"select * from read_parquet('{path}') limit 10").df()
```
